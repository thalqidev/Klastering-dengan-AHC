```python
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist

#dataset
data = {
    'Siswa': ['Siswa1', 'Siswa2', 'Siswa3', 'Siswa4'],
    'Skor Kedisiplinan': [14, 12, 15, 10],
    'Latihan Soal (Kuis)': [8, 7, 9, 6]
}

df = pd.DataFrame(data)

#menghitung jarak dengan manhattan distance
manhattan_distance = pdist(df[['Skor Kedisiplinan', 'Latihan Soal (Kuis)']], metric='cityblock')

#klastering dengan complete linkage
complete_linkage = linkage(manhattan_distance, method='complete')

#visualisasi dendrogram untuk complete linkage
plt.figure(figsize=(10, 5))
dendrogram(complete_linkage, labels=df['Siswa'].values)
plt.title('Dendrogram - Complete Linkage')
plt.xlabel('Siswa')
plt.ylabel('Jarak')
plt.show()

#klastering dengan average linkage
average_linkage = linkage(manhattan_distance, method='average')

# visualisasi dendrogram untuk average linkage
plt.figure(figsize=(10, 5))
dendrogram(average_linkage, labels=df['Siswa'].values)
plt.title('Dendrogram - Average Linkage')
plt.xlabel('Siswa')
plt.ylabel('Jarak')
plt.show()
```
